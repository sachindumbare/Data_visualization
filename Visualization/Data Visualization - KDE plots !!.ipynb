```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
mean = [0,0]
cov = [[0.2,0], [0,3]]
```

```python
x_axis, y_axis = np.random.multivariate_normal(mean, cov, size=40).T
print('x_axis','\n',x_axis)
print('y_axis','\n',y_axis)
```

```python
sns.kdeplot(x_axis)
```

```python
sns.kdeplot(y_axis)
```

```python
sns.kdeplot(x_axis, shade=True)
```

```python
sns.kdeplot(x_axis, shade=True, color='green')
```

```python
sns.kdeplot(y_axis, shade=True, color='purple')
```

```python
sns.kdeplot(x_axis, y_axis)
```

```python
sns.kdeplot(x_axis, y_axis, shade=True)
```

```python
sns.kdeplot(x_axis, y_axis, cmap='RdBu')
```

```python
sns.kdeplot(x_axis, y_axis, cmap='coolwarm')
```

```python
sns.kdeplot(x_axis, y_axis, n_levels=18, cmap='RdBu')
```

```python
sns.kdeplot(x_axis, y_axis, n_levels=35, cmap='coolwarm')
```

```python
sns.kdeplot(y_axis, bw=1.5)
```

```python
sns.kdeplot(y_axis)
```

```python
sns.kdeplot(y_axis, bw=0.1)
```
