# Setting ticks and limits

```python
plt.figure(figsize=(10,5))

x = np.linspace(-6, 6, 200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color='red', linewidth=2.5)
plt.plot(x, sin, color='green', linewidth=3.5)

plt.xlim(-9,9)
plt.ylim(-1.4, 1.4)
```

```python
plt.figure(figsize=(10,5))

x = np.linspace(-6, 6, 200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color='red', linewidth=2.5)
plt.plot(x, sin, color='green', linewidth=3.5)

#plt.xlim(x.min()+0.5, x.max()+0.5)
plt.xlim(x.min()-2, x.max()+2)
plt.ylim(sin.min()*1.3, sin.max()*1.3)
```

```python
plt.figure(figsize=(10,5))

x = np.linspace(-6, 6, 200)

cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color='red', linewidth=2.5)
plt.plot(x, sin, color='green', linewidth=3.5)

plt.xticks([-6,-3,0,1,3,6])
plt.yticks([-1,-.4,0,.5,.8,1])
```

# Adding Text 

```python
# labels, title, text and annot
```

```python
x = np.linspace(-10,10,150)
cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color='red', linewidth=3.0)
```

```python
plt.plot(x, cos, color='red', linewidth=3.0)

plt.xlabel('X-axis')
plt.ylabel('cosx')
plt.title("Graph of x vs cosx")
```

```python
plt.plot(x, cos, color='red', linewidth=3.0)

plt.xlabel('X-axis', fontsize=15, color='green')
plt.ylabel('cosx', fontsize=15, color='blue')
plt.title("Graph of x vs cosx")
plt.grid(True)
```

```python
plt.plot(x, cos, color='red', linewidth=3.0)

plt.xlabel('X-axis', fontsize=15, color='green')
plt.ylabel('cosx', fontsize=15, color='blue')
plt.title("Graph of x vs cosx")
plt.grid(True)

plt.annotate('this is one max value', xy=(0,1), xytext=(5, 0.5),\
             arrowprops=dict(facecolor='black', shrink=0.09))
```

# Line styles and adding text labels

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(-10,10,150)
cos = np.cos(x)
sin = np.sin(x)
```

```python
# :, -., --, -
```

```python
plt.plot(x, cos, color='red', linewidth=3.8, linestyle=':')
```

```python
plt.plot(x, sin, linewidth=3.5, linestyle='-.')
```

```python
plt.plot(x, sin, linewidth=3.5,color='green', linestyle='--')
plt.plot(x, cos, linewidth=3.0,color='red', linestyle='-')
```

```python
plt.plot(x, cos, color='green', linewidth=3.2)

plt.text(-4, -0.3, 'This text will be displayed across our plot', rotation=20, fontsize=14)
```

```python
plt.plot(x, cos, color='green', linewidth=3.2)

plt.text(-7, 0, 'This text will be displayed across our plot', rotation=35, fontsize=14)
```

```python
plt.plot(x, cos, color='green', linewidth=3.2)

plt.text(-8, 0, 'This text will be displayed across our plot', fontsize=14)
```

```python
plt.plot(x, cos, color='green', linewidth=3.2)

plt.text(-4, -0.3, 'This text will be displayed across our plot', rotation=-20, fontsize=14)
```

```python
plt.subplot(211)
plt.plot(x, cos, color='red', linewidth=2.8)
plt.text(-5,-0.5, 'this will run across our subplot', rotation=11, fontsize=14)

plt.subplot(212)
plt.plot(x, sin, color='green', linewidth=3.8)
plt.text(-5,-0.5, 'this will run across our subplot', rotation=-11, fontsize=14)
```

```python
plt.figure(figsize=(10,6))

plt.subplot(121)
plt.plot(x, cos, color='red', linewidth=2.8)
plt.text(-5,-0.5, 'this will run across our subplot', rotation=11, fontsize=14)

plt.subplot(122)
plt.plot(x, sin, color='green', linewidth=3.8)
plt.text(-5,-0.5, 'this will run across our subplot', rotation=-11, fontsize=14)
```

```python
x = np.linspace(-6,6,180)
cos = np.cos(x)
sin = np.sin(x)
```

```python
plt.figure(figsize=(10,7))

plt.subplot(221)
plt.plot(x, cos, color='red', linewidth=2.6, linestyle=':')
plt.title('cosx')
plt.grid(True)
plt.text(-5,-0.3, 'this will run across the subplot', rotation=10, fontsize=15)

plt.subplot(222)
plt.plot(x, sin, color='green', linewidth=3.0, linestyle='-.')
plt.title('sinx')
plt.grid(True)
plt.text(-5,-0.3, 'this will run across the subplot', rotation=-10, fontsize=15)

plt.subplot(223)
plt.plot(x, -cos, color='blue', linewidth=3.5, linestyle='--')
plt.title('-cosx')
plt.grid(True)

plt.subplot(224)
plt.plot(x, -sin, color='purple', linewidth=4.0, linestyle='-')
plt.title('-sinx')
plt.grid(True)
```
